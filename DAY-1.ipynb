# Data Engineering, Data Analytics & Data Science

## Data Engineering
Data Engineering focuses on **building and maintaining data infrastructure**.

- Collects data from users, applications, and external sources  
- Builds and manages data pipelines  
- Performs **ETL / ELT** processes:
  - **Extract** – get data from different sources  
  - **Transform** – clean data and remove unwanted values  
  - **Load** – move data into data warehouses or data lakes  
- Ensures data is reliable, scalable, and high quality  
- Prepares data for Data Analysts and Data Scientists  

---

## Data Analytics
Data Analytics focuses on **analyzing and interpreting data** to find insights.

- Works with cleaned and structured data  
- Answers questions like:
  - What happened?
  - Why did it happen?
- Helps in business decision-making  

### Data Analyst
A Data Analyst is the person who performs data analysis.

- Inspects, cleans, and models data  
- Finds insights (mostly historical and descriptive)  
- Creates reports and dashboards  

**Common tools:**
- SQL  
- Excel  
- Python / R  
- Data visualization tools (Power BI, Tableau, etc.)

---

## Data Science
Data Science focuses on **advanced analytics and predictive modeling**.

- Uses data for:
  - Machine learning  
  - Predictive analytics  
  - Statistical modeling  
- Answers questions like:
  - What will happen next?
  - How can we optimize outcomes?
