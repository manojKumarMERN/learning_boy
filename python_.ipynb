#  Difference Between Parameters and Arguments in Python

> In Python functions:

 **Parameters** are *variables* defined in the **function definition**.  
 **Arguments** are the *actual values* passed to the function when it is **called**.

---

##  1. Parameters - Defined in the Function

Parameters act as **placeholders** for data.  
They exist **only inside the function definition**.

**Example:**
```python
def greet(name):
    print("Hello,", name)
```

##  2. Arguments - Passed to the Function

Arguments are the **real values** you send to the function when calling it.  
We can pass **one or multiple arguments** depending on how the function is defined.

---

###  Example - Passing Multiple Arguments
```python
def add(a, b, c):
    print("a =", a)
    print("b =", b)
    print("c =", c)
    print("Total =", a + b + c)

add(10, 20, 30)

```

### Handling Multiple Arguments with Indexes

f you donâ€™t know how many arguments will be passed,
you can use the special syntax `*args` to accept a variable number of arguments.
This stores all arguments inside a tuple, so you can access them using index values.

```python
def show_numbers(*args):
    print("All numbers:", args)
    print("First number:", args[0])
    print("Second number:", args[1])

show_numbers(10, 20, 30, 40)


```

#  Python Function Arguments â€” Complete Guide

Functions in Python can receive data in multiple ways.  
Understanding different types of **arguments** helps write flexible and reusable code.

---

## ðŸ”¸ What Are Function Arguments?

Arguments are **actual values** passed to a function when it is **called**.  
Parameters are the **variables** defined inside the function that receive those arguments.

---

## ðŸ”¹ 1. Positional Arguments

**Definition:**  
Values are passed **in order**, matching the function parametersâ€™ positions.

**Example:**
```python
def student_info(name, age):
    print(f"Name: {name}, Age: {age}")

# Calling using positional order
student_info("Alice", 15)

```

---

## ðŸ”¹ 2. Keyword Arguments

**Definition:**  
Keyword arguments are passed to a function using the parameter name, so the order doesnâ€™t matter.

**Example:**
```python
def student_info(name, age, grade):
    print(f"Name: {name}, Age: {age}, Grade: {grade}")

# Passing arguments by keyword
student_info(age=15, name="Alice", grade="10th")

```

---

## ðŸ”¹ 3. Default Arguments

**Definition:**  
Default arguments have a predefined value.
If you donâ€™t provide that argument while calling, the default value will be used.

**Example:**
```python
def greet(name, message="Welcome to Python!"):
    print(f"Hello {name}, {message}")

# Using default value
greet("Maddy")

# Overriding default value
greet("Maddy", "Hope you are learning well!")


```

---

## ðŸ”¹ 4. Variable-Length Arguments (`*args`)

**Definition:**  
`*args` allows you to pass any number of positional arguments into a function.
Inside the function, they are stored as a tuple.

**Example:**
```python
def show_songs(*songs):
    print("List of songs:")
    for song in songs:
        print("-", song)

show_songs("Perfect", "Shape of You", "Photograph")

```

---


## ðŸ”¹ 5. Keyword Variable-Length Arguments (`**kwargs`)

**Definition:**  
`**kwargs` allows passing any number of keyword arguments.
Inside the function, they are stored as a dictionary.

**Example:**
```python
def student_profile(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

student_profile(name="Alice", age=15, grade="10th")

```

# âš¡ Lambda Function (Nameless Function)

**Definition:**  
Lambda functions in Python are small, anonymous, **single-expression functions** used for short, simple operations where defining a full function isnâ€™t necessary.

**Syntax:**
```python
lambda arguments: expression

# Traditional function
def square(x):
    return x * x

# Lambda equivalent
square_lambda = lambda x: x * x

print(square_lambda(5))  # Output: 25


```

# âš¡ Map Function 

The `map()` function in Python is used to apply a function to each item in an iterable (like a list or tuple).
It returns a map object (an iterator), which can be converted to a list.

**Syntax:**
```python
map(function, iterable)

numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x * x, numbers))
print(squares)


```
