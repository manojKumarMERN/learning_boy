# NumPy

**NumPy** stands for **Numerical Python**.  
It is a core Python library used for numerical and scientific computing.

## Key Features

- Handles **multi-dimensional arrays** (ndarray)
- Performs **fast mathematical operations** on arrays
- Supports **broadcasting**:
  - Allows operations on arrays of different shapes
  - Avoids explicit loops and improves performance
- Provides tools for **linear algebra**
- Includes **statistical functions**

## Uses of NumPy

- Machine Learning
- Data Engineering
- Data Analysis
- Scientific computing

> NumPy is the foundation library for many Python data and ML libraries like Pandas, SciPy, and Scikit-learn.


```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])
arr2d= arr.reshape(2, 3)

arrflat = arr2d.flatten()

print(f"2D array is {arr2d} \n flatten is {arrflat}" )

```
