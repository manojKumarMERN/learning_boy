# Creating a New Column Using `apply()` and `lambda` (Pandas Only)

## What are we doing?
We are creating a new column from an existing column by applying a condition.
This process is called **feature engineering**.

---

## Example Use Case
From the `Age` column:
- If age < 18 → Child
- If age ≥ 18 → Adult
- If age is missing → Unknown

---

## Why use `apply()` with `lambda`?
- Applies a function to each value in a column
- Useful for simple, readable logic
- Easy to understand for beginners
- No need for NumPy

---

## Basic Syntax
```python
df["new_column"] = df["existing_column"].apply(lambda x: condition)


````python
# titanicDataSet["Age_discribed"] = titanicDataSet["Age"].apply(lambda x: "child" if x < 18 else "adult")

titanicDataSet["Age_discribed"] = titanicDataSet["Age"].apply(
    lambda x: "child" if x < 18 else "adult" if panda.notnull(x) else "unknown"
)

titanicDataSet["Age_discribed"].unique()
titanicDataSet["Age_discribed"].value_counts()
````

# Types of Distribution

Distribution describes **how data values are spread or arranged** in a dataset.

---

## 1. Normal Distribution (Gaussian)

**Description:**
- Data is symmetrically distributed around the mean
- Bell-shaped curve

**Key Points:**
- Mean = Median = Mode
- No skewness

**Examples:**
- Student heights
- Exam scores

---

## 2. Uniform Distribution

**Description:**
- All values occur with equal frequency

**Key Points:**
- Flat distribution
- No peaks

**Examples:**
- Dice roll outcomes
- Random number generation

---

## 3. Right-Skewed Distribution (Positive Skew)

**Description:**
- Most values are small
- Few very large values

**Key Points:**
- Long tail on the right
- Mean > Median

**Examples:**
- Salaries
- House prices

---

## 4. Left-Skewed Distribution (Negative Skew)

**Description:**
- Most values are large
- Few very small values

**Key Points:**
- Long tail on the left
- Mean < Median

**Examples:**
- Exam scores where most students score high

---

## 5. Bimodal Distribution

**Description:**
- Data has two peaks

**Key Points:**
- Two dominant value ranges

**Examples:**
- Heights of men and women combined
- Morning and evening traffic

---

## 6. Multimodal Distribution

**Description:**
- Data has more than two peaks

**Examples:**
- Website traffic across multiple time zones

---

## 7. Exponential Distribution

**Description:**
- Many small values
- Few large values

**Examples:**
- Time between customer arrivals
- Server request gaps

---

## 8. Log-Normal Distribution

**Description:**
- Log of the data follows a normal distribution
- Data is right-skewed

**Examples:**
- Income
- Stock prices
- File sizes

---

## 9. Discrete Distribution

**Description:**
- Data consists of countable values

**Examples:**
- Number of users
- Number of clicks

---

## 10. Continuous Distribution

**Description:**
- Data can take any value within a range

**Examples:**
- Temperature
- Weight
- Time

---

# Quantiles, describe() vs quantile() — Complete Notes

## What are Quantiles?
Quantiles divide a dataset into equal parts to understand data distribution.

Common quantiles:
- Q1 (25%) → 25% of values lie below this
- Q2 (50%) → Median
- Q3 (75%) → 75% of values lie below this

Quantiles help in:
- Understanding data spread
- Detecting skewness
- Identifying outliers

---

## What does `describe()` do?
`describe()` provides a statistical summary of numeric columns.

It includes:
- count
- mean
- std
- min
- **25% (Q1)**
- **50% (Q2 / Median)**
- **75% (Q3)**
- max

So yes:
- 25% → Q1
- 50% → Q2
- 75% → Q3

---

## Why NOT use `describe()` for data processing?
`describe()` is meant for **inspection**, not **logic**.

Limitations:
- Cannot be used easily inside loops
- Cannot automate calculations
- Not suitable for pipelines
- Not used in production code for transformations

Use it only to **understand** the data.

---

## What is `quantile()`?
`quantile()` is used to **programmatically calculate quantiles**.

Examples:
- `quantile(0.25)` → Q1
- `quantile(0.50)` → Median
- `quantile(0.75)` → Q3

It is:
- Precise
- Reusable
- Automation-friendly
- Industry standard

---

## Why Interquartile Range (IQR)?
IQR measures the middle 50% of the data.

Formula:
- IQR = Q3 − Q1

Outlier boundaries:
- Lower Bound = Q1 − 1.5 × IQR
- Upper Bound = Q3 + 1.5 × IQR

Why IQR?
- Robust to extreme values
- Works better than mean-based methods
- Commonly used in real-world datasets

---

## Correct Usage Rule (Very Important)

Use `describe()` when:
- Exploring the dataset
- Understanding distribution
- Manual analysis
- Reporting

Use `quantile()` when:
- Detecting outliers
- Cleaning data
- Writing logic
- Automating workflows
- Production pipelines

---

## Industry Best Practice Workflow

1. Read data
2. Inspect using `info()` and `describe()`
3. Identify numeric columns
4. Use `quantile()` to compute Q1, Q3
5. Calculate IQR
6. Detect outliers
7. Handle outliers (clip / cap / remove)

---

## Interview One-Liner
`describe()` is used for exploratory analysis, while `quantile()` is used for programmatic data processing like outlier detection.

---

- `describe()` → See the problem

````python
# Detect and handle outliers using IQR (Interquartile Range)

# What is Quantile?
# Quantiles help us understand how the data is distributed.
# They split the data into equal parts.
# Example: Q1 is the 25% mark, Q3 is the 75% mark.

# Why Quantile?
# Quantile is used to find the values at specific percentages of the data.
# It helps us understand spread and detect outliers.

# Why Interquartile Range (IQR)?
# IQR measures the middle 50% of the data.
# It is used to detect outliers because it is not affected by extreme values.

# Q1 (25%) is the value below which 25% of data lies.
# Q3 (75%) is the value below which 75% of data lies.
# IQR = Q3 - Q1 (the middle spread of the data)

for col in columnsList:

    # Skip the column if it does not exist in the dataset
    if col not in titanicDataSet.columns:
        continue

    # Apply IQR only on numeric columns (int or float)
    if titanicDataSet[col].dtype != "object":

        # Calculate Q1 and Q3
        q1 = titanicDataSet[col].quantile(0.25)
        q3 = titanicDataSet[col].quantile(0.75)

        # Calculate IQR
        iqr = q3 - q1 

        # Define outlier boundaries
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        # Filter out the outliers
        outliers = titanicDataSet[
            (titanicDataSet[col] < lower_bound) |
            (titanicDataSet[col] > upper_bound)
        ]

        # Why clip?
        # Clip is used to limit values within the boundary.
        # This replaces outliers with the nearest valid value.
        # It helps in keeping the data within a safe range without dropping rows.
        titanicDataSet[col] = titanicDataSet[col].clip(lower=lower_bound, upper=upper_bound)

        # Print the number of outliers found in this column
        print(f"{col} -> Outliers count: {outliers.shape[0]}") 
````
