```python
import pandas as panda

print(panda.__version__)

usersData = {
    "name": ["mano", "uma", "krishna", "vasantha"],
    "age": [20, 21, 22, 23],
    "gender": ["male", "female", "male", "male"],
    "salary": [10000, 20000, 30000, 40000],
    "city": ["bangalore", "chennai", "covai", "trichy"]
}

dataFrame_userData = panda.DataFrame(usersData)
dataFrame_userData.index = ["alpha", "beta", "gamma", "delta"]


filter_age = dataFrame_userData["age"] > 22
sorted_age = dataFrame_userData.sort_values(by="age", ascending=False)

grouping = sorted_age.groupby("gender").agg({"salary": "sum", "age": "mean"})

grouping_pivot = dataFrame_userData.pivot_table(index="name", values="salary", aggfunc="sum")

print(usersData, "\n", dataFrame_userData)

print("filter Data set")
print(dataFrame_userData[filter_age])

print("\n sorted Data set")
print(sorted_age)

print("\n grouping Data set")
print(grouping)

print("\n grouping Pivot table")
print(grouping_pivot)

print("\n grouping Data set")
print(grouping)
grouping_pivot
```

## What is Distribution? (Simple Meaning)

Distribution describes **how values in a dataset are spread** or distributed.




# Types of Distribution

Distribution describes **how data values are spread or arranged** in a dataset.

---

## 1. Normal Distribution (Gaussian)

**Description:**
- Data is symmetrically distributed around the mean
- Bell-shaped curve

**Key Points:**
- Mean = Median = Mode
- No skewness

**Examples:**
- Student heights
- Exam scores

---

## 2. Uniform Distribution

**Description:**
- All values occur with equal frequency

**Key Points:**
- Flat distribution
- No peaks

**Examples:**
- Dice roll outcomes
- Random number generation

---

## 3. Right-Skewed Distribution (Positive Skew)

**Description:**
- Most values are small
- Few very large values

**Key Points:**
- Long tail on the right
- Mean > Median

**Examples:**
- Salaries
- House prices

---

## 4. Left-Skewed Distribution (Negative Skew)

**Description:**
- Most values are large
- Few very small values

**Key Points:**
- Long tail on the left
- Mean < Median

**Examples:**
- Exam scores where most students score high

---

## 5. Bimodal Distribution

**Description:**
- Data has two peaks

**Key Points:**
- Two dominant value ranges

**Examples:**
- Heights of men and women combined
- Morning and evening traffic

---

## 6. Multimodal Distribution

**Description:**
- Data has more than two peaks

**Examples:**
- Website traffic across multiple time zones

---

## 7. Exponential Distribution

**Description:**
- Many small values
- Few large values

**Examples:**
- Time between customer arrivals
- Server request gaps

---

## 8. Log-Normal Distribution

**Description:**
- Log of the data follows a normal distribution
- Data is right-skewed

**Examples:**
- Income
- Stock prices
- File sizes

---

## 9. Discrete Distribution

**Description:**
- Data consists of countable values

**Examples:**
- Number of users
- Number of clicks

---

## 10. Continuous Distribution

**Description:**
- Data can take any value within a range

**Examples:**
- Temperature
- Weight
- Time

---

```python
percentageCalcullate = len(titanicDataSet) / 2

print(percentageCalcullate)
```

```python
for index in range(len(columnsList)):
    column = columnsList[index]

    if titanicDataSet[column].isnull().sum() > percentageCalcullate :
        print(f"{column} have more than {percentageCalcullate} null values {titanicDataSet[column].dtype} ")
    else:
        print(f"{column} have less than {percentageCalcullate} null values {titanicDataSet[column].dtype} ")

```

```python
# we have minimal amount of null values in the age column (float64)
# so we can replace the null values with the mean

titanicDataSet["Age"] = titanicDataSet["Age"].fillna(titanicDataSet["Age"].mean())

titanicDataSet.isnull().sum()
```

```python
# we more than 50% of the null values in the cabin column
# so we can drop the column

# why inplace?
# inplace=True means that the changes will be made to the original dataframe
titanicDataSet.drop("Cabin", axis=1, inplace=True)

titanicDataSet.isnull().sum()
```

```python
# still we have null in Embarked
# we can replace the null values with the median


# why median?
# because Embarked data type is object
titanicDataSet["Embarked"] = titanicDataSet["Embarked"].fillna("Unknown")
```

```python
titanicDataSet.isnull().sum()
```

```python
# Detect and handle outliers using IQR (Interquartile Range)

# What is Quantile?
# Quantiles help us understand how the data is distributed.
# They split the data into equal parts.
# Example: Q1 is the 25% mark, Q3 is the 75% mark.

# Why Quantile?
# Quantile is used to find the values at specific percentages of the data.
# It helps us understand spread and detect outliers.

# Why Interquartile Range (IQR)?
# IQR measures the middle 50% of the data.
# It is used to detect outliers because it is not affected by extreme values.

# Q1 (25%) is the value below which 25% of data lies.
# Q3 (75%) is the value below which 75% of data lies.
# IQR = Q3 - Q1 (the middle spread of the data)

for col in columnsList:

    # Skip the column if it does not exist in the dataset
    if col not in titanicDataSet.columns:
        continue

    # Apply IQR only on numeric columns (int or float)
    if titanicDataSet[col].dtype != "object":

        # Calculate Q1 and Q3
        q1 = titanicDataSet[col].quantile(0.25)
        q3 = titanicDataSet[col].quantile(0.75)

        # Calculate IQR
        iqr = q3 - q1

        # Define outlier boundaries
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        # Filter out the outliers
        outliers = titanicDataSet[
            (titanicDataSet[col] < lower_bound) |
            (titanicDataSet[col] > upper_bound)
        ]

        # Why clip?
        # Clip is used to limit values within the boundary.
        # This replaces outliers with the nearest valid value.
        # It helps in keeping the data within a safe range without dropping rows.
        titanicDataSet[col] = titanicDataSet[col].clip(lower=lower_bound, upper=upper_bound)

        # Print the number of outliers found in this column
        print(f"{col} -> Outliers count: {outliers.shape[0]}")
```
