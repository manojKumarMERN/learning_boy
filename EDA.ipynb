---

## 1. Load the Dataset
- Import required libraries  
- Read the dataset (CSV, Excel, Database, etc.)  
- Store the data in a DataFrame  

---

## 2. Understand the Dataset Structure
- Check total number of rows and columns  
- Review column names  
- Inspect data types of each column  

---

## 3. Preview the Data
- View the first few rows  
- View the last few rows  
- Randomly sample rows for quick inspection  

---

## 4. Check for Duplicate Records
- Identify duplicate rows  
- Count total duplicate entries  
- Remove duplicates if necessary  

---

## 5. Identify Column-wise Data Types
- Separate numeric and categorical columns  
- Identify datetime columns  
- Ensure data types match column meaning  

---

## 6. Check for Missing Values
- Identify columns with null values  
- Count missing values per column  
- Analyze missing value patterns  

---

## 7. Handle Missing Values
- Drop rows or columns if missing data is minimal  
- Fill missing values using:  
  - Mean / Median (numeric)  
  - Mode (categorical)  
  - Custom logic (business rules)  

---

## 8. Statistical Summary (Numeric Columns)
- Minimum value  
- Maximum value  
- Mean  
- Median  
- Quartiles (Q1, Q2, Q3)  
- Standard deviation  

---

## 9. Analyze Data Distribution
- Check value spread  
- Identify skewness  
- Understand central tendency  
- Compare mean vs median  

---

## 10. Detect Outliers
- Use Interquartile Range (IQR)  
- Identify extreme values  
- Decide whether to:  
  - Remove  
  - Cap (clip)  
  - Keep (if valid)  

---

## 11. Analyze Categorical Data
- Count unique categories  
- Frequency distribution  
- Identify dominant categories  
- Detect rare or noisy values  

---

## 12. Feature Engineering
- Create new features from existing columns  
- Combine or transform variables  
- Encode categorical variables if required  

---

## 13. Correlation Analysis
- Identify relationships between numeric features  
- Detect multicollinearity  
- Understand feature importance trends  

---

## 14. Data Validation
- Check logical consistency  
- Verify ranges and constraints  
- Ensure no unexpected values exist  

---

## 15. Data Visualization
- Visualize distributions  
- Compare categories  
- Identify trends and patterns  
- Detect anomalies visually  

---

## 16. Prepare Data for Analytics or ML
- Select features and target variable  
- Encode categorical variables  
- Scale numeric data if required  

---

## 17. Save the Cleaned Dataset
- Export cleaned data  
- Store in CSV or database  
- Ensure reproducibility  

---

## Final Note
> Exploratory Data Analysis (EDA) helps understand data behavior, uncover patterns, detect issues, and build a strong foundation for analytics and machine learning models.

```python
# Distribution of Age column

# Why figure size?
# Controls the width and height of the plot for better readability
plot.figure(figsize=(12, 6))

# Why histplot?
# Histogram is used to visualize the distribution of numeric data
# It shows how frequently values occur  in different ranges (bins)

# Why KDE?
# KDE (Kernel Density Estimation) shows the smooth probability
# density curve of the data, helping understand data distribution shape

# dropna() is used to remove missing values before plotting
sea.histplot(data["Age"].dropna(), kde=True, bins=30,color="green")

# Set plot title and labels for clarity
plot.title("Age Distribution")
plot.xlabel("Age")
plot.ylabel("Count")

# Display the plot
plot.show()
```
